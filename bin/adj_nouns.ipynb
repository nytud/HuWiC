```python
folderpath = '../data/mini_corpus/token_lemma_ana/'
```

```python
import pandas as pd
import re
import os
import sys
from glob import iglob
```

```python
def adj_noun(file):
    adj_noun_regex = '[^/]+/([^/]+)/MN.NOM [^/]+/([^/]+)/FN.'
    file_mn_fn = []
    with open(file) as inp:
        for line in inp:
            for n in re.finditer(rf'{adj_noun_regex}', line):
                adj = n.group(1)
                noun = n.group(2)
                file_mn_fn.append([adj, noun])
    
            
    mn_fns = pd.DataFrame(file_mn_fn)
    try:
        mn_fns.columns = ['mn','fn']
        mn_fns['count'] = 1
        mn_fns_count = mn_fns.groupby(by=['mn', 'fn']).sum().reset_index()
    except ValueError:
        print('Value Error!!')
        print(file)
        mn_fns_count = pd.DataFrame()
    return mn_fns_count
                        
```

```python
input_file = ''
filepaths = [filepath for filepath in iglob(os.path.join(folderpath,'*.txt'))]
filepaths
adj_noun(filepaths[0])

processed_files = []

for file in filepaths:
    print(file)
    new_fn_mn = adj_noun(file)
    processed_files.append(new_fn_mn)

fn_mn_all = pd.concat(processed_files)

```

```python
print(len(fn_mn_all))
fn_mn_all_counts = fn_mn_all.groupby(by=['mn', 'fn']).sum().reset_index().sort_values(by=['mn', 'count'])

print(len(fn_mn_all_counts))

fn_mn_all_counts.to_csv('', sep='\t', index=False)
```
